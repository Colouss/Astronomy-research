```python
import random
import matplotlib.pyplot as plt
#call the necessary arrays and counters
pointin = 0
pointout = 0
inx = []
iny = []
outx = []
outy = []
radius = 3 #The radius of the circle, as well as half the square length
n = 10000 # the number of sampled points we want
for i in range(n):
    #Sample a random x and y value within the constraints of the square
    x = random.uniform(-radius, radius)
    y = random.uniform(-radius, radius)
    r = x**2 + y**2 #calculate the radius of the circle that the randomly regenerated point is on
    if r <= radius**2: #if the radius is smaller or equal than the actual radius,it is within the circle
        inx.append(x) #add the points into the in(side) x and in(side) y arrays
        iny.append(y)
        pointin += 1
    else: #if not, it is outside the circle
        outx.append(x) #add the points into the out(side) x and out(side) y arrays
        outy.append(y)
        pointout += 1
pi = (pointin / pointout) #calculate pi by tallying up how many points are inside the circle with respect to how many are outside
#plot the samples
plt.figure()
plt.scatter(inx, iny, color='blue', label='Inside Circle')
plt.scatter(outx, outy, color='red', label='Outside Circle')
plt.title('Monte Carlo Estimation of Pi')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axis('equal')
plt.show()
#show the result of the estimate
print("Estimated value of pi:", pi)
```

```python
import random
import matplotlib.pyplot as plt
#call the necessary arrays and counters
pointin = 0
pointout = 0
inx = []
iny = []
outx = []
outy = []
def estimatepi(radius, n): #program takes in the radius, and the number of sampled points we want
    pointin = 0
    pointout = 0
    for i in range(n):
        #Sample a random x and y value within the constraints of the square
        x = random.uniform(-radius, radius)
        y = random.uniform(-radius, radius)
        r = x**2 + y**2 #calculate the radius of the circle that the randomly regenerated point is on
        if r <= radius**2: #if the radius is smaller or equal than the actual radius,it is within the circle
            inx.append(x) #add the points into the in(side) x and in(side) y arrays
            iny.append(y)
            pointin += 1
        else: #if not, it is outside the circle
            outx.append(x) #add the points into the out(side) x and out(side) y arrays
            outy.append(y)
            pointout += 1
    pi = (pointin / pointout) #calculate pi by tallying up how many points are inside the circle with respect to how many are outside
    return pi, inx, iny, outx, outy
estimatepi(5, 1000)
#plot the samples
plt.figure()
plt.scatter(inx, iny, color='blue', label='Inside Circle')
plt.scatter(outx, outy, color='red', label='Outside Circle')
plt.title('Monte Carlo Estimation of Pi')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axis('equal')
plt.show()
#show the result of the estimate
print("Estimated value of pi:", pi)
```
